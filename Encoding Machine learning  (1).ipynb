# Odinal encoding 

```python
# make the random data 
dataset=pd.DataFrame({'Size':['s','m','l','xl','m','s','l','xl','xl','m','s','m','l','m','l','xl']})
dataset.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
cor_order=[['s','m','l','xl']]
```

```python
oe=OrdinalEncoder(categories=cor_order)
dataset["New"]=oe.fit_transform(dataset)
```

```python
dataset
```

# Or You can use .map() for doing same thing 

```python
dataset=pd.DataFrame({'Size':['s','m','l','xl','m','s','l','xl','xl','m','s','m','l','m','l','xl']})
dataset.head()
```

```python
cor_order={'s':0,'m':1,'l':2,'xl':3}
```

```python
dataset["New"]=dataset["Size"].map(cor_order)
```
