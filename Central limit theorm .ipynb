```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Generate random number with the help of numpy
dataset=np.random.randint(1,100,size=10000)
dataset[0:5]
```

```python
data=pd.DataFrame(dataset,columns=["Random_no"])
data
```

```python
# Plot kte to see it is normal or not 
plt.figure(figsize=(4,3))
sns.kdeplot(data=data) # The graph is not normal so we can applly central limit theorm.
```

```python
np.random.choice(dataset,size=50)
```

```python
# Take only 10% data from the original data so the 10% of 10000 is 1000
# Lets take 500 samples of sample size 50 
random_sample=[]
random_sample_mean=[]
for i in range(500):
    x=np.random.choice(dataset,size=50)
    y=x.mean()
    random_sample_mean.append(y)
    random_sample.append(x)
print(random_sample_mean)    
    
```

```python
average_mean_dataset=pd.DataFrame(random_sample_mean,columns=["sample_mean"])
average_mean_dataset
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
sns.kdeplot(data=data,ax=axes[0])
axes[0].set_title("Not normal distribution data")
sns.kdeplot(average_mean_dataset,ax=axes[1])
axes[1].set_title("Normal distribution after CLT")
```
