# Feature Selection Technique 

```python
# Their are two types of techniques one is forward elemination and another is backward elemination 
# Import Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Import dataset 
dataset=pd.read_excel(r"C:\Users\ASUS\OneDrive\Documents\Feature selection column.xlsx")
dataset.head(1)
```

```python
dataset.describe()
```

```python
 # Feature selection process is start from here 
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
# Separate the dataset into input and output 
x=dataset.iloc[:,:-1]
y=dataset["Total"]
```

```python
# Import logistic regression 
from sklearn.linear_model import LinearRegression
```

```python
x.columns
```

```python
lr=LinearRegression()
```

```python
fs=SequentialFeatureSelector(lr,k_features=3,forward=True)
```

```python
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

```python
score=[]
feature_name=[]
num=[]
for i in range(1,13):
    fs=SequentialFeatureSelector(lr,k_features=i,forward=True)
    fs.fit(x,y)
    feature_name.append(fs.k_feature_names_)
    score.append(fs.k_score_)
    num.append(i)
    
```

```python
df = pd.DataFrame({"num":num,"feature_name": feature_name,"Score": score})
df

```
