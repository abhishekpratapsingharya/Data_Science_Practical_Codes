```python
# Import libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Import dataset 
dataset=pd.read_excel(r"C:\Users\ASUS\OneDrive\Documents\House hold dataset.xlsx")
dataset.head()
```

```python
# Make the correlation matrics 
dataset.corr()
```

```python
# Their is no highly corrlation between the dataset column 
```

```python
# Scaling the dataset 
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
```

```python
ms.fit(dataset)
```

```python
data=pd.DataFrame(ms.transform(dataset),columns=dataset.columns)
```

```python
data
```

```python
# Split the dataset into x input and y output 
x=data.iloc[:,:-1]
y=data["price"]
```

```python
# Split teh dataset into train and test 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

# Lasso 

```python
from sklearn.linear_model import Lasso
```

```python
la=Lasso(alpha=10)
la.fit(x_train,y_train) # It has internally feature selection option that's why it shows more correct value 
```

```python
la.score(x_test,y_test)*100
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
ri=Ridge(alpha=10)
ri.fit(x_train,y_train)# It does not have any feature selection then it shows better then linear regression 
```

```python
ri.score(x_test,y_test)*100
```
