# Logistic Regression 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Import dataset 
dataset=pd.read_excel(r"C:\Users\ASUS\OneDrive\Desktop\Placed or not.xlsx")
dataset.head(1)
```

```python
sns.scatterplot(x='CGPA',y='TestScore',data=dataset,hue="Placed")
```

```python
# Now Split the dataset into x_input and y_input
x=dataset[["CGPA","TestScore"]]
y=dataset["Placed"]
```

```python
# Import the library for train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Import the library for logistic regression 
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# If i make the prediction line 
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
```
