```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dataset=pd.read_excel(r"C:\Users\ASUS\OneDrive\Desktop\Placed or not.xlsx")
dataset.head()
```

```python
# Scatter Plot 
sns.scatterplot(x=dataset["CGPA"],y=dataset["TestScore"],hue=dataset["Placed"])
```

```python
# Checking the distribution 
fig,axes=plt.subplots(1,2,figsize=(15,3))
sns.kdeplot(dataset["CGPA"],ax=axes[0])
sns.kdeplot(dataset["TestScore"],ax=axes[1]) # Both follows Normal distribution so we can apply Gaussian Naive Bayes.
```

```python
# Split the dataset into x and y 
x=dataset.iloc[:,:-1]
y=dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split The dataset into train and test 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
```

```python
gb.fit(x_train,y_train)
```

```python
gb.score(x_test,y_test)*100
```

```python
# Now lets see what is the decisi0on region plot by the Naive Bayes[Gaussian]
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gb)
```
